```python
from dotenv import load_dotenv
import pandas as pd

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
load_dotenv()
```

```python
movies = pd.read_csv("movies_cleaned.csv")
```

```python
movies["id_description_keywords_genres"].to_csv("id_description_keywords_genres.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader("id_description_keywords_genres.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_movies = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings())
```

```python
query = "A movie about mafia"
docs = db_movies.similarity_search(query, k = 5)
docs
```
